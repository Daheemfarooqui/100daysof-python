# Functions

## Question 7

### Dataset

```python
numbers = [12,15,18,20,25]
```

### Task

Write count_even(numbers) to return the count of even numbers.

Write your solution below.


````python
numbers = [12,15,18,20,25]
def count_even(numbers):
    count = 0
    for i in numbers:
        if i%2==0:
            count += 1
    return count

count_even(numbers)

````

## Question 8

### Dataset

```python
text = 'Data Analyst'
```

### Task

Write count_vowels(text) to return the number of vowels.

Write your solution below.


````python
text = 'Data Analyst'
def count_vowels(text):
    vowel = "AEIOUaeiou"
    count = 0
    for i in text:
        if i in vowel:
            count += 1
    return count
     
count_vowels(text)   
````

## Question 9

### Dataset

```python
numbers = [5,10,15,20]
```

### Task

Write find_max(numbers) without using max().

Write your solution below.


````python
numbers = [5,10,15,200]
def find_max(numbers):
    max_ = 0
    for i in numbers:
        if i > max_:
            max_ = i
    return max_

find_max(numbers)
````
