## Question 3: Students

Dataset: **students.csv**

Task:
- Assign Grade (A/B/C/Fail) and save.

Create the dataset using the code below.

```python
import pandas as pd
from io import StringIO

data = '''Name,Marks
Ali,82
Sara,65
John,35'''

df = pd.read_csv(StringIO(data))

df.loc[df['Marks'] > 80, 'grade'] = 'A'
df.loc[(df['Marks'] > 60) & (df['Marks'] <= 80), 'grade'] = 'B'
df.loc[(df['Marks'] > 40) & (df['Marks'] <= 60), 'grade'] = 'C'
df.loc[df['Marks'] <= 40, 'grade'] = 'Fail'

print(df)
```
