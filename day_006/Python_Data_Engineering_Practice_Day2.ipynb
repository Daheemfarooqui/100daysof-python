# Question 1

## Dataset
```python
import pandas as pd

df = pd.DataFrame({
    "employee":["Ali","Sara","John","David"],
    "department":["Sales","HR","Sales","IT"],
    "salary":[50000,42000,60000,70000]
})
```

### Build an ETL pipeline
- Extract: create `read_data(file_path)`
- Transform: add `bonus = salary*0.10`
- Load: save to `employee_bonus.csv`

Functions to create:
- extract_data()
- transform_data()
- load_data()
- run_pipeline()

````python
# extract
def extract_data(path):
    try:
        df = pd.read_csv(path)
        print('Data extarcted')
        return df

    except:
        print('data not found')


def transform_data(df):
    df['Bonus'] = df['salary']*0.10
    print('transformed')
    return df
    

def load_data(df,OUTPUT_file):
    df.to_csv(OUTPUT_file,index = False)
    print("Report Saved Successfully")

def run_pipeline():
    df = extract_data('Q1/Data.csv')
    df = transform_data(df)
    load_data(df,'Q1/Output.csv')

run_pipeline()
    
````
